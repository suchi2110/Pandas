```python
from matplotlib import pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[30,20,10,40,50]
plt.plot(x,y,marker='s',mfc='y',mec='r',ms=18)
plt.grid(axis='both',c='green',lw=1,ls='dashdot')
plt.show()
```

```python
#lines
import numpy as np
x=np.arange(0,10,2)
y=x**3
y1=x**2
y2=x+x
plt.xlabel('xaxis')
plt.ylabel('x-square')
plt.title('square')
print(x,y)
plt.plot(x,y,ls='dotted',lw=2,marker='o',c='b',ms=13,mfc='blue',mec='green')
plt.plot(x,y1,x,y2,x)
plt.grid(ls='dotted',lw=1,c='green')

plt.show()
```

```python
#to check the font size
x=[1,2,3,4,5]
y=[12,23,34,45,56]
plt.plot(x,y)
f1={'family':'arial','color':'orange', 'size':40}
f2={'family':'arial','color':'green','size':13}
plt.title('oddisgood',fontdict=f1)
plt.xlabel('x-axis',fontdict=f2)
plt.ylabel('y-axis',fontdict=f2)
plt.show()
```

```python
x=[1,2,3,4,5]
y1=[10,30,20,50,40]
y2=[20,30,50,10,40]
y3=[10,50,30,20,40]
y4=[40,20,50,10,30]

plt.suptitle('all-in-one')

plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title('1st')

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title('2nd')

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title('3rd')

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title('4th')

plt.show()
```
