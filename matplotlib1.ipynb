```python
import numpy as np
```

```python
company=['go','am','fb','tw']
revenue=[90,69,79,50]
profit=[20,50,40,66]
#sample=[12,45,66,98]
```

```python
xpos=np.arange(len(company))
xpos
```

```python

plt.ylabel('dollors')
plt.xlabel('company')
plt.title('stock market')
plt.bar(xpos-0.2,revenue,width=0.4,label='revenue')
plt.bar(xpos+0.4,profit,width=0.4,label='profit')
#plt.bar(xpos+0.3,sample,width=0.4,label='sample')
plt.legend(loc='best')
```

```python
year=['2020 Q1','2020 Q2','2020 Q3','2020 Q4']
kent=['55k','34k','22k','48k']
lincoin=['5k','6k','12k','15k']
```

```python
ypos=np.arange(len(year))
ypos
```

```python
plt.xticks(ypos,year)
plt.bar(ypos-0.1,kent,label='money',width=0.3)
plt.bar(ypos+0.1,lincoin,label='lincoin',width=0.2)
plt.legend()
```

```python
x=[90,136,89,27]
x1=[40,2,34,12]
y=['one','two','three','four']
ypos=np.arange(len(y))
plt.xticks(ypos,y)
plt.bar(ypos-0.2,x,width=0.4,label='bars1')
plt.bar(ypos+0.2,x1,width=0.4,label='bars2')
plt.legend()

```

```python
year=['Q1','Q2','Q3','Q4']
money=[505,190,366,87]
dollor=[253,200,12,170]
ypos=np.arange(len(year))
plt.yticks(ypos,year)
plt.barh(ypos-0.1,money,label='money',height=0.3)
plt.barh(ypos+0.1,dollor,label='doll',height=0.2)
plt.legend(loc='best')
```
