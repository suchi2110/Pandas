```python
import pandas as pd
```

```python
india_weather=pd.DataFrame({
    'place' : ['mum','che','kol'],
    'tem' : [32,34,40],
    'hum' :[32,30,29]
})
india_weather
```

```python
japan_weather=pd.DataFrame({
    'place' : ['hiro','tokyo','nag'],
    'tem'  :[24,20,18],
    'hum' :[12,27,23]
    
    
})
japan_weather
```

```python
pd.concat([india_weather,japan_weather])
```

```python
df=pd.concat([india_weather,japan_weather],keys=['india','japan'])
df
```

```python
df.loc['india']
```

```python
df.loc['japan']
```

```python
pd.concat([india_weather,japan_weather],axis=1)
```

```python
tem_df= pd.DataFrame({
    'place' : ['mumbai','che','kol'],
    'hum' : [32,30,29]
} , index=[0,1,2] )
tem_df
```

```python
hum=pd.DataFrame({
    'place' : ['che','mumbai','kol'],
    'tem' : [32,34,40]
},index=[1,0,2])
hum
```

```python
pd.concat([tem_df,hum],axis=1)
```

```python
series=pd.Series(['rain','humidity','cloudy'],name='event')
series
```

```python
df=pd.concat([tem_df,series],axis=1)
df
```

```python
usa=pd.DataFrame({
    'place' : ['nyc','chicago','sd'],
    'tem'   :[23,34,45],
    'hum' :[56,59,67]
})
usa
```

```python
india=pd.DataFrame({
    'place' :['cal','nyc','sd','orlando'],
    'tem' :[34,56,23,33],
    'hum' :[67,69,88,74]
})
india
```

```python
pd.merge(india,usa,on='place',suffixes=('_left','_right'))
```

```python
pd.merge(usa,india,on='place',how='left')
```

```python
pd.merge(usa,india,on='place',how='outer',indicator=True)
```
